```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

tmdb = pd.read_csv("./arquivos/tmdb_5000_movies.csv")
tmdb.head()
```

Ver todas as linguas únicas dos filmes:

```python
tmdb.original_language.unique()
```


As variáveis categóricas são aquelas que não possuem valores quantitativos, mas, ao contrário, são definidas por várias categorias, ou seja, representam uma classificação dos indivíduos. Podem ser nominais ou ordinais.

Quando estamos trabalhando com categorias ordinais, existe uma ordenação entre as categorias. Por exemplo: escolaridade (primeiro, segundo e terceiro graus), estágios de uma doença (inicial, intermediário, terminal), mês de observação (janeiro, fevereiro, ..., dezembro), etc.

O que aprendemos nesta aula:

Importar um CSV;
Identificar o tipo de uma variável observando seu conteúdo;
O que é categórica ordinal;
O que é quantitativa.

Contar quantas vezes uma lingua aparece, é denominada uma SERIE com index pois convertemos pra data frame com o comando to_frame, contendo indice e coluna

```python
tmdb.original_language.value_counts().to_frame()
```

transformando em um dataframe com uma coluna de index e alterando o nome das colunas        

```python
contagem_de_linguas = tmdb.original_language.value_counts().to_frame().reset_index()
contagem_de_linguas.columns = ["lingua_original", "total"]
contagem_de_linguas.head()
```

Criando um gráfico de barras com os dados: (baixo nível)

```python
sns.barplot(x = "lingua_original", y = "total", data = contagem_de_linguas)
```

Alto nível:

```python
sns.catplot(x = "original_language", kind = "count", data = tmdb)
```
