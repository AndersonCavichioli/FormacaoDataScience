```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc('figure', figsize = (20, 10))
```

```python
dados = pd.read_csv('arquivos/aluguel_residencial_novas_colunas.csv', sep = ';')
dados.head(10)
```

```python
grupo_bairro = dados.groupby('Bairro')
grupo_bairro.head(10)
```

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = 'blue')
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor médio do aluguel por bairro', {'fontsize': 22})
```

O mesmo gráfico acima porem com os bairros reduzidos pra ver melhor

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
dados['Bairro'].drop_duplicates()

grupo_bairro = dados.groupby('Bairro')
type(grupo_bairro)
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor médio do aluguel por bairro', {'fontsize': 22})
```

Nesta aula, aprendemos:

Como criar agrupamentos com o groupby()
Estatísticas descritivas com o describe() e o aggregate()
Como renomear as colunas com o rename()
Como fazer gráficos com o pacote Matplotlib
Como criar faixas de valor com o cut()
